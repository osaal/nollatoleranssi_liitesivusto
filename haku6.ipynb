# Haku 6 {#sec-haku-6}

Haku 6 seurasi seuraavia inkluusiokriteerejä:

-   Julkaisuaika: 1.1.2010 - 30.9.2023

-   Metodologia: Empiirinen tutkimus laajasti ymmärrettynä (kvalitatiiviset ja kvantitatiiviset tutkimusasetelmat, toimintatutkimukset joissa kerätään tutkimustietoa, monimenetelmälliset asetelmat)

-   Kohderyhmä: Pelastustoimi ja ensihoito

-   Maantiede: Eurooppa, pois lukien Suomi ja Ahvenanmaa

-   Kieli: Englanti

-   Aineistotyyppi: Vertaisarvioidut tutkimusraportit



+-------------------------------------------------------------+---------------+------------+---------------------------------------------------------------------+-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------+
| Hakukone (tietokannat suluissa)                             | Hakuajankohta | Tulosmäärä | Erityiset rajaukset                                                 | Hakusanat                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                         |
+=============================================================+===============+============+=====================================================================+===================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================================+
| Scopus                                                      | 19.12.2023    | 701        | Tyyppi: Article, Book, Book Chapter, NOT Conference Proceedings     | `TITLE-ABS-KEY ( ( "fire rescue" OR "firefighter" OR "rescue service" OR "emergency respon" OR "emergency personnel" OR {medic} OR "emergency medic" OR ambulance ) AND ( discriminat* OR harass* OR violen* OR "improper behavior" OR equali* OR equit* ) ) AND ( PUBYEAR < 2024 AND PUBYEAR > 2009 ) AND ( DOCTYPE ( ar ) OR DOCTYPE ( bk ) OR DOCTYPE ( ch ) ) AND NOT SRCTYPE ( p ) AND ( LIMIT-TO ( LANGUAGE , "English" ) ) AND ( LIMIT-TO ( AFFILCOUNTRY , "United Kingdom" ) OR LIMIT-TO ( AFFILCOUNTRY , "Undefined" ) OR LIMIT-TO ( AFFILCOUNTRY , "Netherlands" ) OR LIMIT-TO ( AFFILCOUNTRY , "Italy" ) OR LIMIT-TO ( AFFILCOUNTRY , "Spain" ) OR LIMIT-TO ( AFFILCOUNTRY , "Sweden" ) OR LIMIT-TO ( AFFILCOUNTRY , "France" ) OR LIMIT-TO ( AFFILCOUNTRY , "Turkey" ) OR LIMIT-TO ( AFFILCOUNTRY , "Germany" ) OR LIMIT-TO ( AFFILCOUNTRY , "Switzerland" ) OR LIMIT-TO ( AFFILCOUNTRY , "Belgium" ) OR LIMIT-TO ( AFFILCOUNTRY , "Norway" ) OR LIMIT-TO ( AFFILCOUNTRY , "Denmark" ) OR LIMIT-TO ( AFFILCOUNTRY , "Finland" ) OR LIMIT-TO ( AFFILCOUNTRY , "Portugal" ) OR LIMIT-TO ( AFFILCOUNTRY , "Ireland" ) OR LIMIT-TO ( AFFILCOUNTRY , "Greece" ) OR LIMIT-TO ( AFFILCOUNTRY , "Austria" ) OR LIMIT-TO ( AFFILCOUNTRY , "Poland" ) OR LIMIT-TO ( AFFILCOUNTRY , "Czech Republic" ) OR LIMIT-TO ( AFFILCOUNTRY , "Hungary" ) OR LIMIT-TO ( AFFILCOUNTRY , "Romania" ) OR LIMIT-TO ( AFFILCOUNTRY , "Luxembourg" ) OR LIMIT-TO ( AFFILCOUNTRY , "Latvia" ) OR LIMIT-TO ( AFFILCOUNTRY , "Georgia" ) OR LIMIT-TO ( AFFILCOUNTRY , "Slovenia" ) OR LIMIT-TO ( AFFILCOUNTRY , "Serbia" ) OR LIMIT-TO ( AFFILCOUNTRY , "Russian Federation" ) OR LIMIT-TO ( AFFILCOUNTRY , "Lithuania" ) OR LIMIT-TO ( AFFILCOUNTRY , "Iceland" ) OR LIMIT-TO ( AFFILCOUNTRY , "Greenland" ) OR LIMIT-TO ( AFFILCOUNTRY , "Estonia" ) OR LIMIT-TO ( AFFILCOUNTRY , "Cyprus" ) OR LIMIT-TO ( AFFILCOUNTRY , "Croatia" ) OR LIMIT-TO ( AFFILCOUNTRY , "Armenia" ) )` |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Kieli: Englanti                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Maat: Eurooppalaiset maat sekä Undefined[^haku6-5]                  |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Rajaukset toteutettu haun suorittamisen jälkeen.                    |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
+-------------------------------------------------------------+---------------+------------+---------------------------------------------------------------------+-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------+
| Web of Science (WoS Core Collection)                        | 19.12.2023    | 403        | Tyyppi: Article OR Book OR Book Chapter                             | `((TS=(("fire rescue" OR "fire$fighter" OR "rescue service" OR "emergency respon" OR "emergency personnel" OR "medic" OR "emergency medic" OR ambulance) AND (discriminat OR harass* OR violen* OR"improper behavior" OR equali* OR equit*))) AND DOP=(2010-01-01/2023-09-30)) AND DT=(Article OR Book OR Book Chapter)`                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                          |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Kieli: Englanti                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Maat: Eurooppalaiset maat[^haku6-6]                                 |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Kieli- ja maantiederajaukset toteutettu haun suorittamisen jälkeen. |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
+-------------------------------------------------------------+---------------+------------+---------------------------------------------------------------------+-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------+
| Web of Science (MEDLINE)                                    | 19.12.2023    | 1666       | Tyyppi: Journal Article                                             | `(((((TS=(("fire rescue" OR "fire$fighter" OR "rescue service" OR "emergency respon" OR "emergency personnel" OR "medic" OR "emergency medic" OR ambulance) AND (discriminat OR harass* OR violen* OR"improper behavior" OR equali* OR equit*)))) AND LA=(English)) AND DT=(Classical Article OR Clinical Study OR Comparative Study OR Controlled Clinical Trial OR Equivalence Trial OR Journal Article OR Observational Study OR Personal Narrative OR Personal Narratives OR Pragmatic Clinical Trial OR Twin Study OR Validation Studies OR Validation Study))) AND PY=(2010-2023)`                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                          |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Kieli: Englanti                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
+-------------------------------------------------------------+---------------+------------+---------------------------------------------------------------------+-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------+
| ProQuest (Ebook Central, Social Science Premium Collection) | 20.12.2023    | 514        | Vertaisarvioitu: Kyllä                                              | `(TIABSU(("fire rescue" OR "fire?fighter" OR "rescue service" OR "emergency respon" OR "emergency personnel" OR "medic" OR "emergency medic" OR ambulance) AND (discriminat OR harass* OR violen* OR"improper behavio?r" OR equali* OR equit*)) AND PEER(yes) AND PUBDATE(20100101-20230930)) AND (location.exact("United Kingdom UK" OR "England" OR "Europe" OR "Italy" OR "Netherlands" OR "Germany" OR "France" OR "Spain" OR "Sweden" OR "Switzerland" OR "Wales" OR "Denmark" OR "Finland" OR "Norway" OR "Portugal" OR "Belgium" OR "Ireland" OR "Scotland" OR "Georgia" OR "Greece" OR "Austria") AND at.exact("Article" OR "Case Study" OR "Undefined") AND la.exact("ENG") AND stype.exact("Scholarly Journals"))`                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                      |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Lähdetyyppi: Tieteelliset aikakauslehdet                            |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Tyyppi: Article, Case Study, Undefined                              |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Kieli: Englanti                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Maat: Eurooppalaiset maat[^haku6-7]                                 |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
+-------------------------------------------------------------+---------------+------------+---------------------------------------------------------------------+-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------+
| BASE                                                        | 20.12.2023    | 159        | Maat: Eurooppa[^haku6-8]                                            | `("fire rescue" OR firefighter* OR"rescue service" OR "emergency respon" OR "emergency personnel" OR "medic" OR "emergency medic" OR "ambulance") AND (discriminat OR harass* OR violen* OR "improper behavior" OR "equali" OR equit*)`                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                           |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Tyyppi: Book, Article contribution                                  |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            |                                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
|                                                             |               |            | Kieli: Englanti                                                     |                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   |
+-------------------------------------------------------------+---------------+------------+---------------------------------------------------------------------+-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------+
| OpenAlex                                                    | 20.12.2023    | 1049       | Tyyppi: Journal article, Monograph, Report                          | [`https://api.openalex.org/works?search=("fire rescue" OR firefighter OR "rescue service" OR "emergency respon" OR "emergency personnel" OR "emergency medic" OR "ambulance") AND (discriminat OR harass OR violen OR "improper behavior" OR equali OR "non-discriminat" OR equit)&filter=from_publication_date:2010-01-01,to_publication_date:2023-09-30,type_crossref:journal-article|monograph|report&per_page=200&select=title,publication_year,doi&page=1`](https://api.openalex.org/works?search=(%22fire%20rescue%22%20OR%20firefighter%20OR%20%22rescue%20service%22%20OR%20%22emergency%20respon%22%20OR%20%22emergency%20personnel%22%20OR%20%22emergency%20medic%22%20OR%20%22ambulance%22)%20AND%20(discriminat%20OR%20harass%20OR%20violen%20OR%20%22improper%20behavior%22%20OR%20equali%20OR%20%22non-discriminat%22%20OR%20equit)&filter=from_publication_date:2010-01-01,to_publication_date:2023-09-30,type_crossref:journal-article%7Cmonograph%7Creport&per_page=200&select=title,publication_year,doi&page=1)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                |
+-------------------------------------------------------------+---------------+------------+---------------------------------------------------------------------+-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------+

[^haku6-5]: Yhdistyneet kuningaskunnat, Undefined, Hollanti, Italia, Espanja, Ruotsi, Saksa, Sveitsi, Belgia, Norja, Tanska, Suomi, Portugal, Irlanti, Kreikka, Itävalta, Puola, Tsekki, Unkari, Romania, Luxembourg, Latvia, Georgia, Slovenia, Serbia, Venäjä, Liettua, Islanti, Grönlanti, Viro, Kypros, Kroatia ja Armenia. "Suomi" oli virheellisesti lisätty ja mahdolliset tulokset siten poistettu myöhemmässä vaiheessa hakua.

[^haku6-6]: Englanti, Hollanti, Turkki (kahdella kirjoitusmuodolla), Ruotsi, Espanja, Italia, Saksa, Sveitsi, Ranska, Tanska, Skotlanti, Wales, Belgia, Norja, Tsekki, Irlanti, Kreikka, Portugali, Itävalta, Romania, Pohjois-Irlanti, Puola, Bulgaria, Kypros, Unkari, Luxembourg, Venäjä, Slovakia ja Slovenia.

[^haku6-7]: Yhdistyneet kuningaskunnat, Englanti, Eurooppa, Italia, Hollanti, Saksa, Ranska, Espanja, Ruotsi, Sveitsi, Wales, Tanska, Suomi, Norja, Portugali, Belgia, Irlanti, Skotlanti, Georgia, Kreikka ja Itävalta. "Suomi" oli virheellisesti lisätty ja mahdolliset tulokset siten poistettu myöhemmässä vaiheessa hakua.

[^haku6-8]: BASE:n oma rajaus.

Jokaisesta hausta tallennettiin tietolähde, tutkimuksen otsikko, julkaisuvuosi ja tutkimuksen verkko-osoite. Ensisijaisesti haettiin DOI-numeroita, mutta esimerkiksi BASE:n kohdalla osoite voi myös olla muu verkko-osoite.

OpenAlex:in kohdalla tulokset haettiin API-menettelyllä. Maksimitulosten määrä per sivu oli 200 dokumentaation mukaan, joten haku toteutettiin kuusi kertaa nostaen `&page=X` -numeroa jokaisen kerran, `1-6`.

Tulokset ladattiin ensisijaisesti `.csv`- tai `.xlsx`-muotoon jos mahdollista (Scopus, Web of Science ja ProQuest). BASE:n kohdalla lataaminen suoritettiin 100 tulosta kerralla `.json`-muotoon. OpenAlex:in tulokset haettiin 200 tulosta kerralla `.xml`-muotoon.

`.xml`- ja `.json`-muotoiset hakutulokset käännettiin taulukkomuotoon Microsoft Excel -ohjelmassa *Power Query* -toiminnolla.

## Hakutulosten seulonta ja valinta

Hakutulokset seulottiin käyttäen seuraavaa menettelyä:

1.  Kaksoiskappaleet tunnistettiin Microsoft Excelin *Solujen korostussäännöt* -työkalulla. Jokaisen kaksoiskappaleeksi tunnistetun tuloksen kohdalla varmistettiin kaksoiskappaluus, ja säilytettiin yksi tuloksista. Muut merkittiin kaksoiskappaleiksi ja suodatettiin pois taulukosta *Suodata* -työkalulla.
2.  Otsikkojen pohjalta toteutettiin ensimmäinen seulonta. Jos otsikko selkeästi vastasi inkluusiokriteerejä, merkittiin tulokselle arvo 1. Jos otsikko mahdollisesti täytti inkluusiokriteerejä, mutta asiasta ei voinut saada varmuutta, merkittiin tulokselle arvo 2. Jos otsikko selkeästi ei vastannut inkluusiokriteerejä (esim. rikkoi yhden tai useamman kriteerin), jätettiin tyhjä solu. Tyhjät solut suodatettiin sitten pois taulukosta.
3.  Jokaiselle tulokselle, jolla oli merkitty arvo 2 (epävarma) ensimmäisessä seulonnassa, tarkistettiin tiivistelmä tuloksen verkko-osoitteesta. Jos tiivistelmä selkeästi vastasi inkluusiokriteerejä, merkittiin tulokselle arvo 1. Jos tiivistelmä mahdollisesti täytti inkluusiokriteerejä, mutta asiasta ei voinut saada varmuutta, merkittiin tulokselle arvo 2. Jos tiivistelmä selkeästi ei vastannut inkluusiokriteerejä (esim. rikkoi yhden tai useamman kriteerin), jätettiin tyhjä solu. Tyhjät solut suodatettiin sitten pois taulukosta.
4.  Seuraavaksi haettiin kaikki tulokset, jotka saivat arvon 1 ensimmäisessä seulonnassa tai arvot 1 tai 2 toisessa seulonnassa. Jos tuloksen sai haettua, merkittiin arvo 1, ja jos tulosta ei saatu haettua, merkittiin arvo x. Yksi tulos oli kirjaluku, jolloin haettiin kirja ja merkittiin arvo \* kommentin kanssa.
5.  Seuraavaksi tarkistettiin kaikki ne tutkimukset, joille oli merkitty arvo 2 toisessa seulonnassa (epävarma tiivistelmän pohjalta). Tutkija kävi läpi artikkelin menetelmäosiot ja tarkisti, täyttyykö inkluusiokriteerit. Jos kaikki kriteerit täyttyivät, merkittiin arvo 1 (sisällytä kirjallisuuskatsaukseen). Jos osa kriteereistä täyttyivät, merkittiin arvo T (mahdollisesti käytettävissä raportin taustoitukseen, mutta ei kirjallisuuskatsauksessa). Jos kriteerit eivät täyttyneet, tai yksi tai useampi selkeästi rikkoontui, jätettiin tyhjä arvo. Tyhjät solut suodatettiin sitten pois taulukosta.
6.  Viimeiseksi tarkistettiin vielä kaikki tulokset, jotka saivat arvon 1 ensimmäisessä tai toisessa seulonnassa. Tutkija kävi läpi artikkelin menetelmäosiot ja tarkisti, täyttyykö inkluusiokriteerit. Jos kaikki kriteerit täyttyivät, merkittiin arvo 1 (sisällytä kirjallisuuskatsaukseen). Jos osa kriteereistä täyttyivät, merkittiin arvo T (mahdollisesti käytettävissä raportin taustoitukseen, mutta ei kirjallisuuskatsauksessa). Jos kriteerit eivät täyttyneet, tai yksi tai useampi selkeästi rikkoontui, jätettiin tyhjä arvo. Tyhjät solut suodatettiin sitten pois taulukosta.
7.  Valintaprosessin jälkeen jokaisesta tutkimuksesta haettiin seuraavat tiedot:
    1.  Tekijöiden nimet
    2.  Julkaisuvuosi
    3.  Tutkimusjulkaisun otsikko
    4.  Julkaisumuoto (vertaisarvioitu tutkimusartikkeli, organisaatioraportti, kirjateos, kirjan luku, muu)
    5.  Verkko-osoite (ensisijaisesti DOI-numero)
    6.  Tiedonkeruun muoto (kysely, haastattelu, rekisteri, havainnointi, audiovisuaalinen aineisto, muu \[kirjattu avotietona\], useampia muotoja \[kirjattu avotietoina\])
    7.  Kohdepopulaatio (maantieteellinen ja ammatillinen luokittelu)
    8.  Tutkimuksen tiedonkeruun ajankohta (kuukausitasolla, jos tiedossa)
    9.  Otoskoko (yksi tai useampi otos)
    10. Otantamenetelmä (ei tietoa, satunnaistettu otanta, ositettu otanta, ryväsotanta, paneelitutkimus, harkinnanvarainen otanta[^haku6-9], kiintiöotanta, lumipallo-otanta, muu)
    11. Vastaajien yhteydenottotapa (postitse, sähköpostitse, puhelimitse, sosiaalisen verkoston kautta \[esim. kollegat\], organisaatiojohdon kautta, muun organisaation kautta \[esim. liitto\], kasvokkain, muu[^haku6-10])
    12. Vastaajien vastausmuistutusten määrä (jos tiedossa)

[^haku6-9]: Jos tutkimus pyrki koko populaation tutkimiseen, kirjattiin "harkinnanvarainen otanta".

[^haku6-10]: Jos käytettiin useampia yhteydenottomuotoja, merkittiin "muu".

## PRISMA-kulkukaavio

Hakutulosten seulonta tuotti seuraavan PRISMA-protokollan mukaisen kulkukaavion:

\[TÄHÄN KULKUKAAVIO\]

## Valitut tutkimusjulkaisut

\[TÄHÄN LÄHDELUETTELO KAIKISTA MUKAAN VALITUISTA TUTKIMUSJULKAISUISTA\]

## Kokotekstin pohjalta poisseulotut tutkimukset ja syyt

\[TÄHÄN LÄHDELUETTELO KAIKISTA KOKOTEKSTISTÄ POISSEULOTUISTA TUTKIMUKSISTA PÄÄSYYN MUKAAN.\]
